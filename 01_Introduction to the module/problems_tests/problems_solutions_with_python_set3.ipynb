### Question 1: Simulate a projectile motion trajectory given an initial velocity of 50 m/s at an angle of 30°. Plot the trajectory.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define constants
v0 = 50  # Initial velocity in m/s
angle = 30  # Launch angle in degrees
g = 9.81  # Acceleration due to gravity in m/s²

# Step 2: Convert angle to radians
theta = np.radians(angle)

# Step 3: Calculate time of flight, maximum height, and range
t_flight = (2 * v0 * np.sin(theta)) / g
t = np.linspace(0, t_flight, num=500)  # Time steps
x = v0 * np.cos(theta) * t  # Horizontal distance
y = v0 * np.sin(theta) * t - 0.5 * g * t**2  # Vertical distance

# Step 4: Plot the trajectory
plt.plot(x, y)
plt.title("Projectile Motion")
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Distance (m)")
plt.grid()
plt.show()
```

### Question 2: Find the determinant and inverse of a 3x3 matrix:


```python
import numpy as np

# Step 1: Define the matrix
matrix = np.array([[4, 7, 2], [3, 5, 1], [2, 6, 8]])

# Step 2: Calculate the determinant
determinant = np.linalg.det(matrix)

# Step 3: Calculate the inverse
if determinant != 0:
    inverse = np.linalg.inv(matrix)
else:
    inverse = None

# Step 4: Print the results
print("2) Determinant:", round(determinant, 2))
if inverse is not None:
    print("Inverse:\n", inverse)
else:
    print("Inverse does not exist (singular matrix).")
```

### Question 10: Simulate the Lotka-Volterra predator-prey model:


```python
from scipy.integrate import solve_ivp

# Step 1: Define the model
def lotka_volterra(t, z):
    prey, predator = z
    alpha, beta, delta, gamma = 0.1, 0.02, 0.01, 0.1  # Constants
    dprey_dt = alpha * prey - beta * prey * predator
    dpredator_dt = delta * prey * predator - gamma * predator
    return [dprey_dt, dpredator_dt]

# Step 2: Solve the system of equations
t_span = (0, 200)  # Time span
z0 = [40, 9]  # Initial populations (prey, predator)
solution = solve_ivp(lotka_volterra, t_span, z0, t_eval=np.linspace(0, 200, 1000))

# Step 3: Plot the results
plt.plot(solution.t, solution.y[0], label="Prey")
plt.plot(solution.t, solution.y[1], label="Predator")
plt.title("Lotka-Volterra Predator-Prey Model")
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.grid()
plt.show()
```
