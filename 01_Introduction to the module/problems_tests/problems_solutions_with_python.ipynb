```python
import math

# Question (a): Calculate the distance the robot has travelled
# Step 1: Define x and y displacements
x_a, y_a = -0.8, 1.2

# Step 2: Calculate the distance
distance_a = math.sqrt(x_a**2 + y_a**2)

# Step 3: Print the result
print("a) Distance travelled:", round(distance_a, 2), "m")

```

```python

# Question (c): Calculate the travel in x and y for a given distance and heading
# Step 1: Define distance and heading
distance_c = 3.2
heading_c_degrees = 38

# Step 2: Convert heading to radians
heading_c_radians = math.radians(heading_c_degrees)

# Step 3: Calculate x and y displacements
x_c = distance_c * math.cos(heading_c_radians)
y_c = distance_c * math.sin(heading_c_radians)

# Step 4: Print the results
print("c) Travelled in x:", round(x_c, 2), "m, y:", round(y_c, 2), "m")

```

```python
import math

# Question (a): Convert degrees to radians

# Step 1: Define the angles in degrees
angles_in_degrees = [180, 270, -34, 326]

# Step 2: Convert each angle to radians
angles_in_radians = [math.radians(angle) for angle in angles_in_degrees]

# Step 3: Print the results
print("a) Angles in radians:")
for degree, radian in zip(angles_in_degrees, angles_in_radians):
    print(f"{degree} = {round(radian, 2)} radians")


```

```python
import math

# Question (3): Determine the global coordinates of the obstacle

# Step 1: Define the robot's position and orientation
robot_x = 1.2  # Robot's x-coordinate
robot_y = 3.6  # Robot's y-coordinate
robot_heading = 120  # Robot's heading in degrees

# Step 2: Define the obstacle's relative distance and heading
obstacle_distance = 1.4  # Distance to the obstacle
obstacle_heading_relative = -85  # Relative heading of the obstacle in degrees

# Step 3: Calculate the obstacle's global heading
obstacle_global_heading = robot_heading + obstacle_heading_relative

# Step 4: Convert the global heading to radians
obstacle_global_heading_radians = math.radians(obstacle_global_heading)

# Step 5: Calculate the obstacle's global coordinates
obstacle_x = robot_x + obstacle_distance * math.cos(obstacle_global_heading_radians)
obstacle_y = robot_y + obstacle_distance * math.sin(obstacle_global_heading_radians)

# Step 6: Print the global coordinates of the obstacle
print("The obstacle's global coordinates are:")
print(f"x = {round(obstacle_x, 2)} m")
print(f"y = {round(obstacle_y, 2)} m")
```

```python
import math

# Question (4): Calculate the RMSE of the helicopter's control performance

# Step 1: Define the target altitude and sonar readings
target_altitude_mm = 800  # Target altitude in mm
sonar_readings = [85, 82, 79, 75, 72, 77, 84, 86, 83]  # Sonar readings in mm

# Step 2: Calculate the squared errors for each reading
squared_errors = [(reading - target_altitude_mm)**2 for reading in sonar_readings]

# Step 3: Calculate the mean of the squared errors
mean_squared_error = sum(squared_errors) / len(squared_errors)

# Step 4: Calculate the root mean squared error (RMSE)
rmse = math.sqrt(mean_squared_error)

# Step 5: Print the RMSE
print("The RMSE of the helicopter's control performance is:", round(rmse, 2), "mm")
```

```python
# Question (5): Fit a linear trend line using the given data points

# Step 1: Define the data points
current = [42, 36, 41, 22, 34, 28, 38, 24, 26]  # H Bridge Current (mA)
speed = [31, 26, 32, 19, 29, 24, 30, 19, 22]  # Motor Speed (RPM)

# Step 2: Calculate means of current and speed
mean_current = sum(current) / len(current)
mean_speed = sum(speed) / len(speed)

# Step 3: Calculate the slope (m) using the least squares formula
numerator = sum((x - mean_current) * (y - mean_speed) for x, y in zip(current, speed))
denominator = sum((x - mean_current) ** 2 for x in current)
slope = numerator / denominator

# Step 4: Calculate the y-intercept (c)
intercept = mean_speed - slope * mean_current

# Step 5: Print the equation of the line
print(f"The equation of the trend line is: y = {round(slope, 2)}x + {round(intercept, 2)}")

```

```python
import matplotlib.pyplot as plt
```

```python
# Step 5: Generate the trend line
# Define the range of x-values (current) for the line
x_line = range(min(current), max(current) + 1)
# Calculate corresponding y-values using the equation of the line
y_line = [slope * x + intercept for x in x_line]

# Step 6: Plot the data points and the trend line
plt.scatter(current, speed, color="blue", label="Data Points")  # Scatter plot for data points
plt.plot(x_line, y_line, color="red", label="Trend Line")  # Line plot for the trend line

# Step 7: Add labels, legend, and title
plt.xlabel("H Bridge Current (mA)")
plt.ylabel("Motor Speed (RPM)")
plt.title("Linear Trend Line for H Bridge Current vs Motor Speed")
plt.legend()
plt.grid()

# Step 8: Show the plot
plt.show()
```

```python
import math

# Question (6): Calculate the Pearson correlation coefficient

# Step 1: Define the data points
current = [42, 36, 41, 22, 34, 28, 38, 24, 26]  # H Bridge Current (mA)
speed = [31, 26, 32, 19, 29, 24, 30, 19, 22]  # Motor Speed (RPM)

# Step 2: Calculate the mean of the data points
mean_current = sum(current) / len(current)
mean_speed = sum(speed) / len(speed)

# Step 3: Calculate the numerator and denominators of the Pearson formula
numerator = sum((x - mean_current) * (y - mean_speed) for x, y in zip(current, speed))
denominator_x = math.sqrt(sum((x - mean_current) ** 2 for x in current))
denominator_y = math.sqrt(sum((y - mean_speed) ** 2 for y in speed))

# Step 4: Calculate the Pearson correlation coefficient
pearson_coefficient = numerator / (denominator_x * denominator_y)

# Step 5: Print the Pearson correlation coefficient
print("The Pearson correlation coefficient is:", round(pearson_coefficient, 2))
```
