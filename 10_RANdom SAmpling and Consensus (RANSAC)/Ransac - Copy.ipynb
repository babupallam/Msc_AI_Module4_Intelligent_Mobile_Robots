```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn import linear_model, datasets
```

The comparision of the Linear Least Squares Estimate method with RANSAC using the liberary in 'sklearn' package:

```python
n_samples = 1000
n_outliers = 50

#Generate a random regression problem.
x, y, coef = datasets.make_regression(n_samples=n_samples,
                                      n_features=1,
                                      n_informative=1,
                                      noise=10,
                                      coef=True,
                                      random_state=0)#The coefficient of the underlying linear model. It is returned only if coef is True.

# Add outlier data
np.random.seed(0)
x[:n_outliers] = 3 + 0.5 * np.random.normal(size=(n_outliers, 1))
y[:n_outliers] = -3 + 10 * np.random.normal(size=n_outliers)

data = []

for i in range(len(x)):
    data.append([float(x[i]), float(y[i])])

# Fit line using all data
lr = linear_model.LinearRegression()
lr.fit(x, y)

# Robustly fit linear model with RANSAC algorithm in sklearn pakage
ransac = linear_model.RANSACRegressor()
ransac.fit(x, y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

# Predict data of estimated models
line_x = np.arange(x.min(), x.max())[:, np.newaxis]
line_y = lr.predict(line_x)
line_y_ransac = ransac.predict(line_x)

# Compare estimated coefficients
print("Estimated coefficients (true, linear regression, RANSAC):")
print(coef, lr.coef_, ransac.estimator_.coef_)
#print("Estimated coefficients (true, linear regression):")
#print(coef, lr.coef_)
#Ref: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html#sklearn.datasets.make_regression
```

```python
lw = 2
plt.scatter(x[n_outliers:], y[n_outliers:], color='green', marker='.',
            label='Inliers')
plt.scatter(x[:n_outliers], y[:n_outliers], color='red', marker='.',
            label='Outliers')
plt.plot(line_x, line_y, color='navy', linewidth=lw, label='Linear regressor')
plt.plot(line_x, line_y_ransac, color='cornflowerblue', linewidth=lw,
         label='RANSAC regressor')
plt.legend(loc='lower right')
plt.xlabel("X")
plt.ylabel("Response")
plt.show()
```
